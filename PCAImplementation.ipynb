```python
# init libraries
import pandas as pd
import numpy as np
from pathlib import Path
from matplotlib import pyplot as plt

from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler
from sklearn.decomposition import PCA
```

```python
# setup data path and grades
DATA_DIR = Path('Color Data').resolve()
TEA_VARIANTS = ['BOHEA', 'BOP', 'BOPF', 'DUST', 'DUST_II', 'F_I', 'F_II', 'PF', 'PF_II', 'PF_III']
COLOR_MODE = ['RGB', 'HSV', 'LAB']
```

```python
SAMPLE_DIR = Path('Data_NEW').resolve()
total_samples = {}

for variant in TEA_VARIANTS:
    sample_dir = SAMPLE_DIR / variant
    csv_files = sorted(sample_dir.glob('*.jpg'))
    
    total_samples[variant] = len(csv_files)
    
print(total_samples)
```

```python
# load data
all_samples = []
sample_label = []

for variant in TEA_VARIANTS:
    for mode in COLOR_MODE:
        data_dir = DATA_DIR / variant / f"{variant}_{mode}"
        csv_files = sorted(list(data_dir.glob('*.csv')))
        
        for csv_file in csv_files:
            filename = data_dir / csv_file
            df_sample = pd.read_csv(csv_file)
            feature_vector = df_sample.to_numpy().flatten()
            
            all_samples.append(feature_vector)
            sample_label.append(variant)
            
data_df = pd.DataFrame(all_samples, index=sample_label)

print(f"data shape: {data_df.shape}")
print(data_df.head())
```
